```julia
using Plots
using Random
using LinearAlgebra
```

```julia
Random.seed!(0)
# Gerando dados fictícios para teste
n = 30
x = range(0, 1, length=n)

x1(x) = x
x2(x) = x^2
x3(x) = x^3
x4(x) = x^4
x5(x) = x^5

y =  x1.(x) + x2.(x) + x3.(x) + x4.(x) + x5.(x) + rand(n)*0.1#ruído
scatter(x, y, c=:lightblue, ms=3, leg=false)
```

```julia
function vandermonde(x,y,grau)
    n,=size(y)
    V=zeros(n,grau+1)
    for i=1:n #linhas
        for j=1:(grau+1)
            V[i,j]=x[i]^(j-1)
        end
    end
    return V
end
```

```julia
function regressão(x,y,grau)
    V=vandermonde(x,y,grau)
    c=V\y #mínimos quadrados
    return c
end
```

```julia
c=regressão(x,y,5)
```

```julia
grau_5(x)= c[1] + c[2]*x + c[3]*x^2 + c[4]*x^3 + c[5]*x^4 + c[6]*x^5
```

```julia
scatter(x, y, c=:lightblue, ms=3, leg=false)
plot!(grau_5)
```
